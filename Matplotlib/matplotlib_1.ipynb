```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = np.arange(0,101,10)
y_data = x_data**2
noise = np.random.randint(-10,10,11)
y_data = np.add(y_data, noise)
```

```python
plt.plot(x_data,y_data)
plt.show
```

Basic costumization

```python
fig =plt.figure(1, figsize=(5,5))
plt.plot(x_data,y_data)
plt.show
```

```python
fig =plt.figure(1, figsize=(5,5))
plt.plot(x_data,y_data, marker='o', markersize = '10', mfc = 'white', linestyle = '--', label = 'data set 1')
plt.legend(loc = 'lower right')
plt.show()
```

```python
#two sets of data in the same plot

fig =plt.figure(1, figsize=(5,5))

plt.plot(x_data,y_data, marker='o', markersize = '10', mfc = 'white', linestyle = '--', label = 'data set 1')

plt.plot(x_data,y_data + 500, marker='s', markersize = '10', mfc = 'white', linestyle = '-', label = 'data set 1')

plt.legend(loc = 'lower right')
plt.show()
```

Logaritmic plot

```python
fig =plt.figure(1, figsize=(5,5))

plt.semilogy(x_data,y_data, marker='o', markersize = '10', mfc = 'white', linestyle = '--', label = 'data set 1')

plt.semilogy(x_data,y_data + 500, marker='s', markersize = '10', mfc = 'white', linestyle = '-', label = 'data set 1')

plt.legend(loc = 'lower right')
plt.show()
```

Math labels

```python
fig =plt.figure(1, figsize=(5,5))

plt.semilogy(x_data,y_data, marker='o', markersize = '10', mfc = 'white', linestyle = '--', label = '$y=x^2$')

plt.semilogy(x_data,y_data + 500, marker='s', markersize = '10', mfc = 'white', linestyle = '-', label = '$y=x^2+500$')

plt.legend(loc = 'lower right')
plt.show()
```

Axis labels

```python
fig =plt.figure(1, figsize=(5,5))

plt.semilogy(x_data,y_data, marker='o', markersize = '10', mfc = 'white', linestyle = '--', label = '$y=x^2$')

plt.semilogy(x_data,y_data + 500, marker='s', markersize = '10', mfc = 'white', linestyle = '-', label = '$y=x^2+500$')

plt.xlabel('x variable')
plt.ylabel('y label')
plt.legend(loc = 'lower right')
plt.show()
```
